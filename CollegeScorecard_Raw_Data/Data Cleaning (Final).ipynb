### Filtering dataset cases
Now that we've imported the data, we will need to filter the dataset to only include colleges that mainly offer bachelor's degrees. 

```python
# Only include schools predominantly offering bachelor's degrees, and filter out graduate and law schools included
# accidentally.
bach=Data14[(Data14['PREDDEG']==3) & (Data14['SCH_DEG']==3) & (~Data14['INSTNM'].str.contains("Graduate")) 
             &(~Data14['INSTNM'].str.contains("Law School"))].copy()
# drop columns that only have null values
bach=bach.drop(bach[['PREDDEG','SCH_DEG']],axis=1)
bach=bach.dropna(axis=1,how='all')
```

### Renaming columns
We then need to rename the columns so that they're more understandable. This requires extracting the original and dev-friendly names from rows of the cohort map where the 2014-15 column isn't null, and shortening the dev-friendly names.

```python
# create devnames-data, which has the original variable names and the corresponding developer-friendly rename
devnames_data=cohort[cohort['2014-15'].notnull()][['name','dev_name']]
# various abbreviations/renames to shorten the names and keep them understandable
devnames_data.dev_name=devnames_data.dev_name.str.replace('percentile','pctl')
devnames_data.dev_name=devnames_data.dev_name.str.replace('_scores','')
devnames_data.dev_name=devnames_data.dev_name.str.replace('title_iv','t_iv')
devnames_data.dev_name=devnames_data.dev_name.str.replace('yrs_after_entry','yrs_AE')
devnames_data.dev_name=devnames_data.dev_name.str.replace('working_not_enrolled','WNE')
devnames_data.dev_name=devnames_data.dev_name.str.replace('greater_than_','>')
devnames_data.dev_name=devnames_data.dev_name.str.replace('less_than_','<')
devnames_data.dev_name=devnames_data.dev_name.str.replace('percent_','pct')
devnames_data.dev_name=devnames_data.dev_name.str.replace('-plus','+')
devnames_data.dev_name=devnames_data.dev_name.str.replace('by_income_level','by_income')
devnames_data.dev_name=devnames_data.dev_name.str.replace('repayment','repay')
devnames_data.dev_name=devnames_data.dev_name.str.replace('year_declining','yr_declining')
devnames_data.dev_name=devnames_data.dev_name.str.replace('first_generation','1st_gen')
devnames_data.dev_name=devnames_data.dev_name.str.replace('transfer','transf')
devnames_data.dev_name=devnames_data.dev_name.str.replace('110001plus','110001+')
devnames_data.dev_name=devnames_data.dev_name.str.replace('dependent_students','dependent')
devnames_data.dev_name=devnames_data.dev_name.str.replace('percentage','pctg')
devnames_data.dev_name=devnames_data.dev_name.str.replace('program.bachelors','bach')
devnames_data.dev_name=devnames_data.dev_name.str.replace('first_gen','1st_gen')
devnames_data.dev_name=devnames_data.dev_name.str.replace('completion','compl')
```

### Column Trimming
Now that the columns have been made neater and more consistent, we'll get rid of many of them - specifically, ones where overall variables are disaggregated by demographic groups. The reason for this is because there's a lot of data that is missing or not reported when down on that stratified level, and because reducing the ambitious scope of the original project proposal makes cleaning and analyzing the data much easier. We'll also be removing cohort counts, which are the sample sizes of the cohorts used to calculate the student-related variables in the dataset, since those aren't relevant to our analysis.

#### Net Price
Variables in the dataset relating to net price are split between public and private, due to calculations for net price differing between public and private institutions. None of the 'public' columns overlap with their corresponding 'private' columns, so we can merge the two average net price columns into one column. However, since we don't care about sub-demographic data, we'll remove columns where net price is broken down by income level. We'll also remove Title IV income cohort counts, which are also split by institution type.

```python
# merging public and private columns together
renamed.loc[:,'avg_net_price.public'].fillna(renamed.loc[:,'avg_net_price.private'],inplace=True)
# rename columns
combining=renamed.rename(columns={'avg_net_price.public': 'avg_net_price'})
# delete private columns, net price by income columns, & t_iv income cohort count columns
combining=combining.drop(['avg_net_price.private'],axis=1)
combining=combining.drop([col for col in combining if 'net_price' in col and 'by_income' in col],axis=1)
combining=combining.drop([col for col in combining if 't_iv' in col and 'by_income' in col],axis=1)
```

#### Filling in null values for completion rate
We begin by looking at the cases where `compl_rate_4yr_150nt` is either null or 0. Still, a two-year difference isn't massive enough to be an issue for our purposes. Using the `.info()` method on specific slices of the DataFrame, we can determine whether replacing null and 0 values with corresponding Title IV variables would decrease the percentage of missing/invalid values to below 1%. 

```python
#looking at rows where compl_rate_4yr_150nt is null
cc[cc['compl_rate_4yr_150nt'].isnull()][
    ['t_iv.completed_by.6yrs']].info()
```

#### Retention Rate
The retention rate is the percentage of students who return to the institution after their first year. As with before, there are two variables (one general and one for Title IV students) related to student retention over a similar time period: `retention_rate.four_year.full_time` and `t_iv.still_enrolled_by.2yrs`. The former is for the 2013-2014 enrollment cohort, while the latter is for the 2012-2013 enrollment cohort. Still, we will assume that the actual retention rate doesn't change dramatically between one year and the next. Although the number of 0 values is below 1%, it doesn't make sense for any collge to have no students returning the next year, so we need to clean those cases as well. Again, we use the `.info()` method to determine whether filling in null and 0 values with corresponding Title IV values would make this column complete enough.

```python
cc[cc['retention_rate.four_year.full_time'].isnull()][['retention_rate.four_year.full_time',
                                                        't_iv.still_enrolled_by.2yrs']].info()
```

```python
cc[(cc['retention_rate.four_year.full_time']==0)& (cc['t_iv.still_enrolled_by.2yrs']!=0)][[
    't_iv.still_enrolled_by.2yrs']].info()
```

Since only 12 null variables and 1 zero-valued variable would remain in the retention rate column, it's viable to fill in those values with the corresponding Title IV values.

```python
cc.loc[:,'retention_rate.four_year.full_time'].fillna(cc.loc[:,'t_iv.still_enrolled_by.2yrs'],inplace=True)
cc.loc[:,'retention_rate.four_year.full_time'].replace(0,cc.loc[:,'t_iv.still_enrolled_by.2yrs'],inplace=True)
```

#### Imputation by Group
For our tuition variables, it makes sense to group the cases by the `ownership` variable; private and for-profit colleges tend to be more expensive than public colleges, which are partially government-funded and cheaper for students residing in the same state as the college.

```python
college_type=cc.groupby(['ownership'])
netprice_mean_bygroup=college_type['avg_net_price'].mean()
instate_mean_bygroup=college_type['tuition.in_state'].mean()
outofstate_mean_bygroup=college_type['tuition.out_of_state'].mean()
pd.concat([netprice_mean_bygroup,instate_mean_bygroup,outofstate_mean_bygroup],axis=1)
```

The `impute_mean()` function is adapted from a similar function in DataCamp's [Manipulating DataFrames with pandas](https://campus.datacamp.com/courses/manipulating-dataframes-with-pandas/grouping-data?ex=11) course.

```python
def impute_mean(series):
    return series.fillna(series.mean())

cc['avg_net_price']=college_type['avg_net_price'].transform(impute_mean)
cc['tuition.in_state']=college_type['tuition.in_state'].transform(impute_mean)
cc['tuition.out_of_state']=college_type['tuition.out_of_state'].transform(impute_mean)
```

#### Filling null values with mean
For two of our columns, we fill the null values with the mean of the entire column.

```python
#filling na with mean for certain variables
cc['faculty_salary'].fillna(np.mean(cc['faculty_salary']),inplace=True)
cc['t_iv.withdrawn_by.3yrs'].fillna(np.mean(cc['t_iv.withdrawn_by.3yrs']),inplace=True)
```

#### Share of first-time, full-time students
There are two columns in the dataset pertaining to the share of full-time versus part-time students at the college: `share_first.time_full.time` and `part_time_share`. The total share of students at a college can be thought of as first-time full-time share + part-time share + non-first-time full-time share, and should equal 1. Subtracting the part-time share from 1 (provided that the part-time share isn't 0) can get a good upper estimate of the share of full-time and first-time students. 

```python
cond1=(cc['share_first.time_full.time']==0) &  (cc['part_time_share']!=0)
cond2=(cc['share_first.time_full.time'].isnull())\
&(cc['part_time_share'].notnull())\
&(cc['part_time_share']!=0)
cond3=(cc['share_first.time_full.time']==0) &  (cc['part_time_share']==0)

cc['share_first.time_full.time']=cc['share_first.time_full.time'].mask(cond2,1-cc['part_time_share'])
cc['share_first.time_full.time']=cc['share_first.time_full.time'].mask(cond1,1-cc['part_time_share'])
cc[(cc['share_first.time_full.time'].isnull()) | (cc['share_first.time_full.time']==0)]['share_first.time_full.time']
```

#### Filling in null values for categorical variable
Finally, we fill in missing values for `open_admissions_policy`, a categorical variable desribing whether the college accepts anyone who applies. We will add a new category '0' to indicate that the admissions policy of the college is unknown.

```python
cc['open_admissions_policy']=cc['open_admissions_policy'].cat.add_categories([0]) #0 indicates unknown value
cc['open_admissions_policy'].fillna(0, inplace=True)
```
