```python
# reading in the cohort map, which describes which columns are present in which years
cohort_map=pd.read_excel('CollegeScorecardDataDictionary.xlsx',sheet_name=4)
```

```python
# reading in the data dictionary, which has descriptions of each variable 
data_dict = pd.read_excel('CollegeScorecardDataDictionary.xlsx', sheet_name=3)
```

```python
# get all valid columns and store them in non_null_columns
non_null_cols = data_dict[data_dict['developer-friendly name'].notnull()]

# extract developer-friendly name and dev category as colnames dataframe
devnames_cohort = non_null_cols[['developer-friendly name','dev-category', 'NAME OF DATA ELEMENT','VARIABLE NAME']]
devnames_cohort=devnames_cohort.reset_index(drop=True)

# append dev_name and dev_category from colnames to cohort_map
cohort_merge=pd.merge(cohort_map,devnames_cohort, left_on='Variable Name',right_on='VARIABLE NAME')
```

```python
# Removing columns from before 2010 and making cohort map easier to read
cohort=pd.concat([cohort_merge.iloc[:,0:1],cohort_merge.iloc[:,15:25]],axis=1)
cohort.columns=['name','2010-11','2011-12','2012-13','2013-14','2014-15','2015-16','2016-17','dev_name','dev_category','info']
cohort
```

* mean post-graduate earnings
* loan repayment rate

Because the 2014-2015 dataset is the most recent one containing information on earnings stratified by certain subgroups, I have decided to use that year's dataset for this project.

```python
Data14 = pd.read_csv("MERGED2014_15_PP.csv", na_values=['PrivacySuppressed'],  dtype={'ZIP':str, 'ALIAS':str})
```

```python
Data14.count()
```

Now that we've imported the data, we will need to filter the dataset to only include colleges that mainly offer bachelor's degrees. 

```python
bach=Data14[(Data14['PREDDEG']==3) & (~Data14['INSTNM'].str.contains("Graduate")) 
             &(~Data14['INSTNM'].str.contains("Law School"))]
bach=bach.dropna(axis=1,how='all')
```

```python
devnames_data=cohort[cohort['2014-15'].notnull()][['name','dev_name']]
devnames_data.dev_name=devnames_data.dev_name.str.replace('percentile','pctl')
devnames_data.dev_name=devnames_data.dev_name.str.replace('_scores','')
devnames_data.dev_name=devnames_data.dev_name.str.replace('title_iv','tiv')
devnames_data.dev_name=devnames_data.dev_name.str.replace('yrs_after_entry','yrs_AE')
devnames_data.dev_name=devnames_data.dev_name.str.replace('working_not_enrolled','WNE')
devnames_data.dev_name=devnames_data.dev_name.str.replace('greater_than_','>')
devnames_data.dev_name=devnames_data.dev_name.str.replace('less_than_','<')
devnames_data.dev_name=devnames_data.dev_name.str.replace('percent_','pct')
devnames_data.dev_name=devnames_data.dev_name.str.replace('-plus','+')
devnames_data.dev_name=devnames_data.dev_name.str.replace('by_income_level','by_income')
devnames_data.dev_name=devnames_data.dev_name.str.replace('repayment','repay')
devnames_data.dev_name=devnames_data.dev_name.str.replace('first_generation','1st_gen')
devnames_data.dev_name=devnames_data.dev_name.str.replace('transfer','transf')
devnames_data.dev_name=devnames_data.dev_name.str.replace('110001plus','110001+')
devnames_data.dev_name=devnames_data.dev_name.str.replace('dependent_students','dependent')
devnames_data.dev_name=devnames_data.dev_name.str.replace('percentage','pctg')
devnames_data.dev_name=devnames_data.dev_name.str.replace('bachelors','bach')
devnames_data.dev_name=devnames_data.dev_name.str.replace('first_gen','1st_gen')
devnames_data.dev_name=devnames_data.dev_name.str.replace('completion','compl')
collist=pd.DataFrame(bach.columns)
collist.columns=['name']
newcols=list(devnames_data.merge(collist).iloc[:,1])
bach.head()
```

```python
bach.columns=newcols
bach.head()
bach.count()
keepcols=[c for c in bach.columns if 'certificate_lt' not in c]
keepcols=[c for c in keepcols if 'program.assoc' not in c]

len(bach[(bach['tiv.public.all'].notnull()) & 
        (bach['tiv.private.all'].notnull())])
```

```python
#remove columns for number of degrees if those columns aren't related to bachelor's degrees
cleaned=bach[keepcols]
# combine public and private columns together
cleaned.loc[:,'avg_net_price.public']=cleaned['avg_net_price.public'].combine_first(cleaned['avg_net_price.private'])
cleaned.loc[:,'net_price.public.by_income.0-30000']=cleaned[
    'net_price.public.by_income.0-30000'].combine_first(cleaned['net_price.private.by_income.0-30000'])
cleaned.loc[:,'net_price.public.by_income.30001-48000']=cleaned[
    'net_price.public.by_income.30001-48000'].combine_first(cleaned['net_price.private.by_income.30001-48000'])
cleaned.loc[:,'net_price.public.by_income.48001-75000']=cleaned[
    'net_price.public.by_income.48001-75000'].combine_first(cleaned['net_price.private.by_income.48001-75000'])
cleaned.loc[:,'net_price.public.by_income.75001-110000']=cleaned[
    'net_price.public.by_income.75001-110000'].combine_first(cleaned['net_price.private.by_income.75001-110000'])
cleaned.loc[:,'net_price.public.by_income.110001+']=cleaned[
    'net_price.public.by_income.110001+'].combine_first(cleaned['net_price.private.by_income.110001+'])
cleaned.loc[:,'net_price.public.by_income.0-48000']=cleaned[
    'net_price.public.by_income.0-48000'].combine_first(cleaned['net_price.private.by_income.0-48000'])
cleaned.loc[:,'net_price.public.by_income.30001-75000']=cleaned[
    'net_price.public.by_income.30001-75000'].combine_first(cleaned['net_price.private.by_income.30001-75000'])
cleaned.loc[:,'net_price.public.by_income.75000+']=cleaned[
    'net_price.public.by_income.75000+'].combine_first(cleaned['net_price.private.by_income.75000+'])
cleaned.loc[:,'tiv.public.all']=cleaned['tiv.public.all'].combine_first(cleaned['tiv.private.all'])
cleaned.loc[:,'tiv.public.by_income.0-30000']=cleaned[
    'tiv.public.by_income.0-30000'].combine_first(cleaned['tiv.private.by_income.0-30000'])
cleaned.loc[:,'tiv.public.by_income.30001-48000']=cleaned[
    'tiv.public.by_income.30001-48000'].combine_first(cleaned['tiv.private.by_income.30001-48000'])
cleaned.loc[:,'tiv.public.by_income.48001-75000']=cleaned[
    'tiv.public.by_income.48001-75000'].combine_first(cleaned['tiv.private.by_income.48001-75000'])
cleaned.loc[:,'tiv.public.by_income.75001-110000']=cleaned[
    'tiv.public.by_income.75001-110000'].combine_first(cleaned['tiv.private.by_income.75001-110000'])
cleaned.loc[:,'tiv.public.by_income.110001+']=cleaned[
    'tiv.public.by_income.110001+'].combine_first(cleaned['tiv.private.by_income.110001+'])

# rename columns
cleaned.rename(columns={'avg_net_price.public': 'avg_net_price',
                        'net_price.public.by_income.0-30000': 'net_price.by_income.0-30000',
                        'net_price.public.by_income.30001-48000': 'net_price.by_income.30001-48000',
                        'net_price.public.by_income.48001-75000': 'net_price.by_income.48001-75000',
                        'net_price.public.by_income.75001-110000': 'net_price.by_income.75001-110000',
                        'net_price.public.by_income.110001+': 'net_price.by_income.110001+',
                        'net_price.public.by_income.0-48000': 'net_price.by_income.0-48000',
                        'net_price.public.by_income.30001-75000': 'net_price.by_income.30001-75000',
                        'net_price.public.by_income.75000+': 'net_price.by_income.75000+',
                        'tiv.public.all':'tiv.all', 'tiv.public.by_income.0-30000':'tiv.by_income.0-30000',
                        'tiv.public.by_income.30001-48000':'tiv.by_income.30001-48000',
                        'tiv.public.by_income.48001-75000':'tiv.by_income.48001-75000',
                        'tiv.public.by_income.75001-110000':'tiv.by_income.75001-110000',
                        'tiv.public.by_income.110001+':'tiv.by_income.110001+'
                       }, inplace=True)
# delete private columns
cleaned=cleaned.drop(['avg_net_price.private','net_price.private.by_income.0-30000',
                      'net_price.private.by_income.30001-48000','net_price.private.by_income.48001-75000',
                      'net_price.private.by_income.75001-110000','net_price.private.by_income.110001+', 
                      'net_price.private.by_income.0-48000','net_price.private.by_income.30001-75000',
                      'net_price.private.by_income.75000+','tiv.private.all','tiv.private.by_income.0-30000',
                      'tiv.private.by_income.30001-48000','tiv.private.by_income.48001-75000',
                      'tiv.private.by_income.75001-110000','tiv.private.by_income.110001+'],axis=1)
```
